```python
import boto3
```

```python
account_id = boto3.client("sts").get_caller_identity().get("Account")
region = boto3.Session().region_name
ecr_repository = "strand_agents_pilot"
tag = ":latest"
processing_repository_uri = "{}.dkr.ecr.{}.amazonaws.com/{}".format(
    account_id, region, ecr_repository + tag
)
```

```python
# Login to ECR
!aws ecr get-login-password --region {region} | docker login --username AWS --password-stdin {account_id}.dkr.ecr.{region}.amazonaws.com
```

```python
# Create ECR repository
!aws ecr create-repository --repository-name $ecr_repository --region {region}
```

```python
# build and tag docker image
# !docker build -t {processing_repository_uri} ../../
!docker build --platform linux/arm64 -t {processing_repository_uri} ../../
```

```python
# push the image to repository
!docker push $processing_repository_uri
```

```python
# Verify the image
!aws ecr describe-images --repository-name {ecr_repository} --region {region}
```
